# **Univariate Analysis**

**Salary Distribution**

```python
sns.histplot(df['Salary'])
plt.title('Salary Distribution')
plt.show()
```

Observation: The salary distribution is right-skewed

```python
n = ['Salary', '10percentage', '12percentage', 'CollegeTier', 'collegeGPA',
                     'GraduationYear', 'English', 'Logical', 'Quant', 'Domain',
                     'ComputerProgramming', 'ElectronicsAndSemicon', 'ComputerScience',
                     'MechanicalEngg', 'ElectricalEngg', 'TelecomEngg', 'CivilEngg',
                     'conscientiousness', 'agreeableness', 'extraversion', 'nueroticism',
                     'openess_to_experience']

plt.figure(figsize=(20, 15))
for i, column in enumerate(n):
    plt.subplot(5, 5, i+1)
    sns.boxplot(x=df[column])
    plt.title(f'Boxplot of {column}')
    plt.xlabel(column)
    plt.tight_layout()

for i in range(len(n), 25):
    plt.subplot(5, 5, i+1)
    plt.axis('off')

plt.show()
```

```python
sns.boxplot(df['Salary'])
plt.title('Salary Boxplot')
plt.show()
```

There are few outliers

```python
df = df[df['Salary'] <= 800000]
```

```python
sns.boxplot(df['Salary'])
plt.title('Salary Boxplot(outliers removed)')
plt.show()
```

Observation: Outliers removed, salary range is more representative.

**Specialization Distribution**

**Gender Distribution**

# **Bivariate Analysis**
**Quant Score vs Salary**


```python
sns.scatterplot(x='Quant', y='Salary', data=df)
plt.title('Quant Score vs Salary')
plt.show()
```

**Pairplot**


```python
cols = ['Salary', 'Quant', 'English', 'Logical']
sns.pairplot(df[cols], height=2.5)
plt.show()
```

**Correlation Heatmap**


```python
corr_matrix = df[['Salary', 'Quant', 'English', 'Logical']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Heatmap')
plt.show()
```


**Average Salary by Degree**


```python
avg_salaries = df.groupby('Degree')['Salary'].mean().reset_index()
plt.figure(figsize=(8, 6))
sns.barplot(x='Degree', y='Salary', data=avg_salaries)
plt.title('Average Salary by Degree')
plt.xlabel('Degree')
plt.ylabel('Average Salary')
plt.xticks(rotation=90)
plt.show()
```

**Specialization vs Salary**


```python
sns.boxplot(x='Specialization', y='Salary', data=df)
plt.title('Specialization vs Salary')
plt.xticks(rotation=90)
plt.show()
```

# **Research Question**
-Times of India article dated Jan 18, 2019 states that âAfter doing your Computer Science Engineering if you take up jobs as a Programming Analyst, Software Engineer, Hardware Engineer and Associate Engineer you can earn up to 2.5-3 lakhs as a fresh graduate.â

-Is there a relationship between gender and specialization? (i.e. Does the preference of Specialisation depend on the Gender?)


```python
avg_salaries_degree = df.groupby('Degree')['Salary'].mean()
print(avg_salaries_degree)
```

```python
relevent_jobs = ["programmer analyst","software engineer", "hardware engineer" , "associate engineer"]
df1 = df[(df["Degree"]=="B.Tech/B.E.") & (df["Specialization"] == "computer science & engineering") & (df["Designation"].isin(relevent_jobs))]

plt.figure(figsize=(10, 10))
sns.boxplot(x='Designation', y='Salary', data=df1)
plt.title('Boxplot: Designation vs Salary')
plt.xlabel('Designation')
plt.ylabel('Salary')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Specialization', hue='Gender', data=df)
plt.title('Gender vs Specialization')
plt.xlabel('Specialization')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

# **Conclusion**

The analysis of Computer Science & Engineering graduates' salaries reveals that Programmer Analysts, Software Engineers, and Associate Engineers earn salaries ranging from lower to upper bounds.  Partially supporting the Times of India's claim, Programmer Analysts and Software Engineers' salaries fall within or close to the 2.5-3 lakhs per annum range. Associate Engineers' salaries vary, but some may fall within this range. Further analysis with expanded data is recommended to confirmÂ theseÂ findings.
